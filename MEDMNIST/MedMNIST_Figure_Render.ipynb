```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

# These are all the pre-computed robustness and 
# Accuracy values from teh evaulation scripts

p_std_acc = 0.881
p_rob_acc = 0.854
p_std_inp_atk = 0.99
p_rob_inp_atk = 1.00
p_std_inp_rob = 0.00
p_rob_inp_rob = 1.00
p_rob_mod_atk = 1.00
p_rob_mod_rob = 1.00
p_std_mod_atk = 0.98
p_std_mod_rob = 0.00


d_std_acc = 0.72
d_rob_acc = 0.67
d_std_inp_atk = 0.66
d_std_inp_rob = 0.00
d_rob_inp_atk = 0.98
d_rob_inp_rob = 0.78
d_rob_mod_atk = 1.00
d_rob_mod_rob = 0.89
d_std_mod_atk = 0.04
d_std_mod_rob = 0.00

b_std_acc = 0.88
b_rob_acc = 0.68

b_std_inp_atk = 0.67
b_std_inp_rob = 0.00

b_rob_inp_atk = 0.93
b_rob_inp_rob = 0.72

b_rob_mod_atk = 0.72
b_rob_mod_rob = 0.39

b_std_mod_atk = 0.06
b_std_mod_rob = 0.00
```

```python


sns.set_context('poster')

labels = ['Accuracy', 'Inp. Attack Rob.', 'Mod. Attack Rob.', 'Inp. Cert.', 'Mod. Cert.']
std_heights = [d_std_acc, d_std_inp_atk, d_std_mod_atk, d_std_inp_rob, d_std_mod_rob]
rob_heights = [d_rob_acc, d_rob_inp_atk, d_rob_mod_atk, d_rob_inp_rob, d_rob_mod_rob]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars
##85604d
fig, ax = plt.subplots(figsize=(15,8), dpi=100)
rects1 = ax.bar(x - width/2, std_heights, width, label='Standard', color = "#bf8c71", edgecolor='k')
rects2 = ax.bar(x + width/2, rob_heights, width, label='Grad. Cert.', hatch='/', color = "#bf8c71", edgecolor='k')



# Add some text for labels, title and custom x-axis tick labels, etc.
#ax.set_ylabel('Scores')
ax.set_title('Accuracy and Robustness on DermaMNIST')
ax.set_xticks(x, labels)
#ax.legend()

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)

fig.tight_layout()

plt.ylim((0,1.15))
#plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), ncol=2)
plt.show()
```
